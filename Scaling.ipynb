#**Scaling**
  - Scaling is nothing but bringing all the features to a similar scale

##Types of Scaling

     --->min-max scaling
     --->standardization(if you have outliers

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("/content/scalingminmax_student_data (1).csv")
df
```

```python
scalar = MinMaxScaler()
df["Income"]=scalar.fit_transform(df[["Income"]])
print(df)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("/content/scalingminmax_student_data (1).csv")
scalar = StandardScaler()
df["Income"]=scalar.fit_transform(df[["Income"]])
print(df)
```
