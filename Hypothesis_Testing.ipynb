```python
#a school principle is saying my college average is 80
import numpy as np
from scipy.stats import ttest_1samp
#taking sample
marks=[82,85,78,90,88,79,84,86,87,83]
#ho=80
t_statistic,p_value=ttest_1samp(marks,80)
print(np.mean(marks))
print("tstatisic : ",t_statistic)
print("p_value : ",p_value)
alpha=0.05
if p_value<alpha:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
