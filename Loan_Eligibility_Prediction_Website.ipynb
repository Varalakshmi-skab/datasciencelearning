```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets for datascience/loandataset.csv')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['LoanAmount'].fillna(df['LoanAmount'].median(), inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].median(), inplace=True)
df['Credit_History'].fillna(df['Credit_History'].median(), inplace=True)
```

```python
df.isnull().sum()
```

```python
categorical_cols = ['Gender','Married','Dependents','Self_Employed']
for columns in categorical_cols:
    df[columns].fillna(df[columns].mode()[0], inplace=True)
```

```python
le = LabelEncoder()
cat_cols_to_encode = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']
for columns in cat_cols_to_encode:
    df[columns] = le.fit_transform(df[columns])
```

```python
X = df.drop(['Loan_Status', 'Loan_ID'],axis=1)
Y = df['Loan_Status']
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
print("Accuracy")
print(accuracy_score(Y_test,y_pred))
print("Confusion Matrix")
print(confusion_matrix(Y_test,y_pred))
print("Precision Score")
print(precision_score(Y_test,y_pred))
print("Recall Score")
print(recall_score(Y_test,y_pred))
print("F1 Score")
print(f1_score(Y_test,y_pred))
```

```python
import pickle
import gradio as gr
import numpy as np
model = pickle.load(open("MLmodel.pkl", "rb"))
def predict_loan(gender, married, dependents, education, self_employed,
                 applicant_income, coapplicant_income, loan_amount,
                 loan_amount_term, credit_history, property_area):
    input_data = np.array([[gender, married, dependents, education, self_employed,
                            applicant_income, coapplicant_income, loan_amount,
                            loan_amount_term, credit_history, property_area]])
    pred = model.predict(input_data)[0]
    if pred == 1:
        return "Loan Approved"
    else:
        return "Loan Rejected"
app = gr.Interface(
    fn=predict_loan,
    inputs=[
        gr.Dropdown(["Female", "Male"], type="index", label="Gender"),
        gr.Dropdown(["No", "Yes"], type="index", label="Married"),
        gr.Dropdown(["0", "1", "2", "3+"], type="index", label="Dependents"),
        gr.Dropdown(["Graduate", "Not Graduate"], type="index", label="Education"),
        gr.Dropdown(["No", "Yes"], type="index", label="Self Employed"),
        gr.Number(label="Applicant Income"),
        gr.Number(label="Coapplicant Income"),
        gr.Number(label="Loan Amount"),
        gr.Number(label="Loan Amount Term (in days)"),
        gr.Dropdown([0.0, 1.0], label="Credit History"),
        gr.Dropdown(["Rural", "Semiurban", "Urban"], type="index", label="Property Area")
    ],
    outputs=gr.Textbox(label="Loan Status Prediction"),
    title="Loan Eligibility Prediction Website",
    description="Enter the applicant's details below to predict whether their loan will be approved or rejected."
)
app.launch(share=True)
```
