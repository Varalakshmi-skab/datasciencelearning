```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets for datascience/netflix_titles.csv")
df.head()
```

```python
df['director'].fillna('Unknown',inplace=True)
```

```python
df['cast'].fillna('Unknown',inplace=True)
```

```python
df['country'].fillna('Unknown',inplace=True)
```

```python
df['rating'].fillna('Unknown',inplace=True)
```

```python
df['description'].fillna('Unknown',inplace=True)
df['date_added'].fillna(pd.Timestamp('2000-01-01'), inplace=True)
```

#part two

```python
df['date_added']=pd.to_datetime(df['date_added'], errors='coerce')
df['AddedYear']=df['date_added'].dt.year
```

```python
df['AddedMonth']=df['date_added'].dt.month
```

```python
df['AddedDay']=df['date_added'].dt.day
```

```python
df['AddedWeekDay']=df['date_added'].dt.day_name()
```

```python
df['AddedQuarter']=df['date_added'].dt.quarter
```

```python
current_year=pd.Timestamp.now().year
df['ContentAge']=current_year-df['release_year']
```

```python
df['YearsToNetflix']=df['AddedYear']-df['release_year']
```

```python
df['TitleLength']=df['title'].str.len()
```

```python
df.head()
```

```python
df['TitleWordCount']=df['title'].str.split().str.len()
```

```python
df['LongTitle']=df['TitleWordCount']>3
```

```python
df['DescriptionLength']=df['description'].str.len()
```

```python
df
```

```python
df['DescriptionWordCount']=df['description'].str.split().str.len()
```

```python
df
```

```python
df['ActorCount']=df['cast'].str.split(',').str.len().fillna(0)
```

```python
df['CountryCount']=df['country'].str.split(',').str.len().fillna(0)
```

```python
df
```

```python
df['GenreCount']=df['listed_in'].str.split(',').str.len().fillna(0)
```

```python
df['MultiGenre']=df['GenreCount']>1
```

```python
df
```

```python
df['IsMovie']=df['type']=='Movie'
```

```python
df['IsTVShow']=df['type']=='TV Show'
```

```python
df.head()
```

```python
df['RecentContent']=df['release_year']>=2020
```

```python
df['ClassicContent']=df['release_year']<2000
```
