#**data time handling** is the process of converting date and time into a format that computers can uunderstand

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets for datascience/transaction_data (1).csv")
df
```

```python
df.info()
```

```python
df["transaction_date"]=pd.to_datetime(df["transaction_date"])
df
df.info()
```

```python
df["transaction_time"]=pd.to_datetime(df["transaction_time"])
df
df.info()
```

```python
df["transaction_datetime"]=pd.to_datetime(df["transaction_datetime"])
df
df.info()
```

```python
df
```

```python
df["year"]=df["transaction_date"].dt.year
df
```

```python
df["month"]=df["transaction_date"].dt.month
df
```

```python
df["day"]=df["transaction_date"].dt.day
df
```

```python
df["month_name"]=df["transaction_date"].dt.month_name()
df
```

```python
df["day_name"]=df["transaction_date"].dt.day_name()
df
```

#Extract quarter

```python
df["quarter"]=df["transaction_date"].dt.quarter
df
```

```python
df["hour"]=df["transaction_datetime"].dt.hour
df["minutes"]=df["transaction_datetime"].dt.minute
df["seconds"]=df["transaction_datetime"].dt.second
df
```
