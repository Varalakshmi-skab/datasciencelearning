```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets for datascience/economic_index.csv")
```

```python
df.isnull().sum()
df.duplicated().sum()
df.describe()
df.info()
```

```python
df.corr()
```

```python
sns.pairplot(df)
```

```python
X=df[["interest_rate","unemployment_rate"]]
X.shape
Y=df["index_price"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
print(model.intercept_)
print(model.coef_)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
y_pred_acc=r2_score(y_test,y_pred)
print(y_pred_acc)#testing accuracy
```

```python
y_train_pred=model.predict(X_train)
y_train_acc=r2_score(y_train,y_train_pred)
print(y_train_acc)
```

```python
mae = mean_absolute_error(y_test, y_pred)
print("MAE :", mae)
mse = mean_squared_error(y_test, y_pred)
print("MSE :", mse)
rmse = np.sqrt(mse)
print("RMSE :", rmse)
```

```python
import pickle
pickle.dump(model,open("MLRmodel.pkl","wb"))
pickle.dump(scaler,open("MLRscaling.pkl","wb"))
```

```python
import gradio as gr
import pickle
model=pickle.load(open("/content/drive/MyDrive/Datasets for datascience/MLRmodel.pkl","rb"))
scaler=pickle.load(open("/content/drive/MyDrive/Datasets for datascience/MLRscaling.pkl","rb"))
```

```python
def predict_index(interest_rate,unemployment_rate):
  data=[[interest_rate,unemployment_rate]]
  scaled_data=scaler.transform(data)
  pred=model.predict(scaled_data)
  return int(pred[0])
app=gr.Interface(
    fn=predict_index,
    inputs=[gr.Number(label="interest_rate"),
            gr.Number(label="unemployment_rate")],
    outputs=gr.Number(label="predicted index price"),
    title="economic index price prediction"
)
app.launch(share=True
)
```
