#**Time Series Data**
    ---> Time Series Data is the data collected over time
    ---> example : stock price
    ---> in normal data row order is not important
    ---> in time series data we need particular data

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets for datascience/stock_price_data.csv")
df
```

```python
df.info()
```

```python
df.head()
```

```python
#step1 : find datatypes : if datatypes is not in proper format just change the format
```

```python
#converting date column
df["Date"]=pd.to_datetime(df["Date"])
df.info()
df
```

```python
#sort the values
dfd=df.sort_values("Date")
dfd
```

```python
#set date as index
dfd.set_index("Date",inplace=True)
dfd
```

below code meaning:

H-hour

D-day

W-week

M-month

Y-year


```python
#we can create missing dates too : we use "asfreq" for that
dfd=dfd.asfreq("D")
dfd
```

```python
#handling missing values
# day1 - 100
# day2 -
# day3 -
# day4 - 160
# 160-100=60/3=20=>missing values :120,140
# interpolation : it means filling missing values by using the values before and after them and estimating what the missing value should be
dfd["Stock_Price"]=dfd["Stock_Price"].interpolate()
dfd["Volume"]=dfd["Volume"].interpolate()
dfd
```

```python
#other methods
#forward fill : replaces a missing value with the previous available value
#(ffill)
#backward fill : replaces a missing value with the next available value
#(bfill)
```

```python
dfd
```

```python
#converting date features
dfd["Year"]=dfd.index.year
dfd["Month"]=dfd.index.month
dfd["Day"]=dfd.index.day
dfd["Dayofweek"]=dfd.index.dayofweek
# monday-0
# tuesday-1
# wednesday-2
# thursday-3
# friday-4
# saturday-5
# sunday-6
dfd
```

```python
dfd["Weekend"]=(dfd.index.dayofweek>=5).astype(int)
dfd
```

```python
#lag feature
# a lag feature means it uses past values to predict future values
dfd["Lag1"]=dfd["Stock_Price"].shift(1)
dfd
dfd["Lag2"]=dfd["Stock_Price"].shift(2)
dfd
dfd["Moving_average"]=dfd["Stock_Price"].rolling(3).mean()
dfd
dfd["Max_roll"]=dfd["Stock_Price"].rolling(3).max()
dfd
```
