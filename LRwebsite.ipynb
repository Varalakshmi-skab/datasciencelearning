```python
import pickle
import gradio as gr
model=pickle.load(open("/content/drive/MyDrive/Datasets for datascience/linear_regression_model.pkl",
                       "rb"))
scalar=pickle.load(open("/content/drive/MyDrive/Datasets for datascience/scaler.pkl",
                        "rb"))
def predict_height(weight):
  scaled=scalar.transform([[weight]])
  prediction=model.predict(scaled)
  return prediction[0]
app=gr.Interface(
    fn=predict_height,
    inputs=gr.Number(label="enter weight"),
    outputs=gr.Number(label="Predicted Height"),
    title="Weight Height Prediction"
)
app.launch()
```
