```python
import pandas as pd
import numpy as np
```

#series

```python
#series
arr=np.array([10,20,30,40,50])
s=pd.Series(arr)
print(s)
print(s[0])
```

```python
marks=pd.Series(
    [10,20,30],
    index=["math","phy","chem"]
)
print(marks)
print(marks["math"])
#shape: it gives rows and colomns👇
print(marks.shape)
#size:it gives no. of elements👇
print(marks.size)
print(marks.ndim)
print(marks.dtype)
print(marks.index)
print(marks.values)
```

#applyling operations using loops👇👇

```python
#=> 2,4,68,14
numbers=[1,2,34,7]
result=[]
for i in numbers:
  result.append(i*2)
print(result)
```

#***vectorized*** ***operation*** means applying one operation to all elements automatically without writing loops👇👇

```python
numbers=[1,2,34,7]
arr=np.array(numbers)
print(arr*2)
```

#if i want to know the missing values.i will use isnull() method

```python
data=pd.Series([10,20,30,40,np.nan,50])
# [10,20,30,40,np.nan,50]
print(data.isnull())
print(data.fillna(0))
```

#***DataFrame***

```python
#dictionaries={}
data={
    "name": ["st1","st2","st3","st4"],
    "marks": [10,20,30,40],
    "city": ["c1","c2","c3","c4"]
}
df=pd.DataFrame(data)
print(df)
print(df.shape)
print(df.size)
print(df.columns)
print(df.index)
print(df.ndim)
print(df.info)
```

#***create*** a new column

```python
df["grade"]=["b","b","a","a"]
print(df)
```

#***updating*** a column

```python
df["marks"]=df["marks"]+20
print(df)
```

```python
df.drop("city",axis=1)
```
