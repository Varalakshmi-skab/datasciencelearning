```python
#series
arr=np.array([10,20,30,40,50])
s=pd.Series(arr)
print(s)
print(s[0])
```

```python
marks=pd.Series(
    [10,20,30],
    index=["math","phy","chem"]
)
print(marks)
print(marks["math"])
#shape: it gives rows and colomns👇
print(marks.shape)
#size:it gives no. of elements👇
print(marks.size)
print(marks.ndim)
print(marks.dtype)
print(marks.index)
print(marks.values)
```

#**Imputation** : filling missing values in datasets by using some methods or logic

-->Mean

-->Median

-->Mode

```python
import seaborn as sns
df=sns.load_dataset("titanic")
df
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df["age"]
```

```python
df.columns
```

```python
df["Age_Mean"]=df["age"].fillna(df["age"].mean())
print(df["Age_Mean"])
```

```python
df[["Age_Mean","age"]]
```

```python
df["Age_Median"]=df["age"].fillna(df["age"].median())
print(df["Age_Median"])
```

```python
df[["Age_Median","age"]]
```

```python
df["Age_Mode"]=df["age"].fillna(df["age"].mode())
print(df["Age_Mode"])
```

```python
print(df["age"].mode())
```

```python
df
```

```python
df["age"]=df["age"].fillna(df["age"].mean())
```
