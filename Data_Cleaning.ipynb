```python
import pandas as pd
```

```python
df=pd.read_csv("/content/employee_data (1).csv")
df
```

```python
df["Age"]=df["Age"].fillna(df["Age"].mean())
df["Employee Name"]=df["Employee Name"].fillna("Unknown")
```

```python
df.duplicated().sum()
df=df.drop_duplicates()
df.head()
```

```python
df["Employee Name"]=df["Employee Name"].str.strip()
df["Department"]=df["Department"].str.strip()
```

```python
df["Department"]=df["Department"].replace(
    {
        "IT":"Information Technology",
        "it":"Information Technology"
    }
)
df
```

```python
df.columns=(
    df.columns.str.lower().str.replace(" ","_")
)
df
```

```python
df['join_date'] = pd.to_datetime(df['join_date'], format='mixed')
print(df['join_date'].dtype)
```
