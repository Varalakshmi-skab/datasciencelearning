```python
import numpy as np
data=[2,4,5,6,7,8,9,100]
#step1 q1,q3
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
print(Q1,Q3)
#step2 iqr
IQR=Q3-Q1
print(IQR)
#step3 and 4
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print(lower)
print(upper)
outliers=[]
for i in data:
  if i<lower or i>upper:
    outliers.append(i)
print(outliers)
cleaned_outliers=[]
for i in data:
  if i>lower and i<upper:
    cleaned_outliers.append(i)
print(cleaned_outliers)
```
