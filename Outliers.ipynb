```python
import numpy as np
data=[2,4,5,6,7,8,9,100]
#step1 q1,q3
Q1=np.percentile(data,25)
Q3=np.percentile(data,75)
print(Q1,Q3)
#step2 iqr
IQR=Q3-Q1
print(IQR)
#step3 and 4
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
print(lower)
print(upper)
outliers=[]
for i in data:
  if i<lower or i>upper:
    outliers.append(i)
print(outliers)
cleaned_outliers=[]
for i in data:
  if i>lower and i<upper:
    cleaned_outliers.append(i)
print(cleaned_outliers)
```

#Statistical Measures

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets for datascience/student_marks_outlier_dataset.csv")
df
```

```python
#Central tendency
print("Mean",df["Marks"].mean())
print("Median",df["Marks"].median())
print("Mode",df["Marks"].mode())
```

#Dispersions

```python
print("Variance",df["Marks"].var())
print("Standard deviation",df["Marks"].std())
```

#Quantiles

```python
Q1=df["Marks"].quantile(0.25)
Q2=df["Marks"].quantile(0.50)
Q3=df["Marks"].quantile(0.75)
print("Quantile 1 :",Q1)
print("Quantile 2 :",Q2)
print("Quantile 3 :",Q3)
```

#Skewness

```python
skewness=df["Marks"].skew()
print("Skewness :",skewness)
if skewness>0:
  print("Positively Skewd")
elif skewness<0:
  print("Negatively Skwed")
else:
  print("Symmetric Distribution")
```

#Outlier Detection

```python
Q1=df["Marks"].quantile(0.25)
Q3=df["Marks"].quantile(0.75)
IQR=Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
outliers=[]
for i in df["Marks"]:
  if i<lower or i>upper:
    outliers.append(i)
print(outliers)
df.drop(19,inplace=True)
df
```
