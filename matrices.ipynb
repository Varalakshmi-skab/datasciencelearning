```python
import numpy as np
```

#matrix operations

```python
mat1=np.array([
    [1,2],
    [3,4]
])
mat2=np.array([
    [5,6],
    [7,8]
])
```

#addition operation

```python
result=mat1+mat2
print(result)
```

#subtraction operation

```python
result=mat2-mat1
print(result)
```

#multiplication operation

```python
result=mat1*mat2
print(result)
```

#methods of matrix multiplication

```python
result=np.matmul(mat1,mat2)
print(result)
result=mat1 @mat2
print(result)
result=np.dot(mat1,mat2)
print(result)
```

```python
result=mat1/mat2
print(result)
```
