#aggregate functions are functions used to summerize

-->multiple data values into a single meaningful result

-->sum,max,min,mean

```python
import pandas as pd
```

```python
df=pd.read_excel("/content/drive/MyDrive/Datasets for datascience/student_dataset_20rows.xlsx")
```

```python
df.head()
```

```python
#sum : it will calculates sum all the data in that column
print(df["Marks"].sum())
```

```python
#It will calculate the mean
print(df["Marks"].mean())
```

```python
#It will calculate the median
print(df["Marks"].median())
```

```python
#It will calculate the minimum
print(df["Marks"].min())
```

```python
#It will calculate the maximum
print(df["Marks"].max())
```

```python
print(df["Marks"].count())
```

```python
print(df["Marks"].std())
```

```python
print(df["Marks"].var())
```

```python
print(df["Marks"].describe())
```

```python
print(df["Marks"].agg(["sum","mean"]))
```
