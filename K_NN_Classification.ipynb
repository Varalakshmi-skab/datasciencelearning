```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets for datascience/knnclassification.csv")
df.head()
```

```python
X=df.drop("Loan_Status",axis=1)
X
Y=df["Loan_Status"]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
X_train=scalar.fit_transform(X_train)
X_test=scalar.transform(X_test)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,Y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
confusion_matrix=confusion_matrix(Y_test,y_pred)
print(confusion_matrix)
accuracy_score=accuracy_score(Y_test,y_pred)
print(accuracy_score)
precision_score=precision_score(Y_test,y_pred,pos_label="Approved")
print(precision_score)
recall_score=recall_score(Y_test,y_pred,pos_label="Approved")
print(recall_score)
f1_score=f1_score(Y_test,y_pred,pos_label="Approved")
print(f1_score)
```
