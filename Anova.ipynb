#Anova : it is analysis of variance

    --->Anova helps us compare multiple groups and chek whether their average values are significantly different from each other

```python
from scipy.stats import f_oneway
method_a=[70,72,68,75,74]
method_b=[80,82,78,85,81]
method_c=[90,92,88,95,91]
#f_stat tells us whether the grps are significantly different from each other
f_stat,p_value=f_oneway(method_a,method_b,method_c)
print("f-statistic:",f_stat)
print("p-value:",p_value)
if p_value<0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
from scipy.stats import chisquare
observed = [20,10,5,5]
expected = [10,10,10,10]
chi_stat,p_value=chisquare(observed,expected)
print("chi-statistic:",chi_stat)
print("p-value:",p_value)
if p_value<0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
from statsmodels.stats.proportion import proportions_ztest
sucess=[180,120]
samples=[500,500]
z_stat,p_value=proportions_ztest(sucess,samples)
print("z-statistic:",z_stat)
print("p-value:",p_value)
if p_value<0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("/content/drive/MyDrive/Datasets for datascience/flight_price.xlsx")
df.head()
```

```python
df['Date_of_Journey']=pd.to_datetime(df['Date_of_Journey'],format="%d/%m/%Y")
df['Journey_Day']=df['Date_of_Journey'].dt.day
df['Journey_Month']=df['Date_of_Journey'].dt.month
df['Journey_Year']=df['Date_of_Journey'].dt.year
```

```python
df.head()
```

```python
df['Dep_Time']=pd.to_datetime(df['Dep_Time'])
df['dept_hour']=df['Dep_Time'].dt.hour
df['dept_min']=df['Dep_Time'].dt.minute
df.head()
```

```python
df['Total_Stops']=df['Total_Stops'].replace({
    'non-stop':0,
    '1 stop':1,
    '2 stops':2,
    '3 stops':3,
    '4 stops':4
})
df.head()
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df['Airline']=encoder.fit_transform(df['Airline'])
df['Source']=encoder.fit_transform(df['Source'])
df['Destination']=encoder.fit_transform(df['Destination'])
```
