#Anova : it is analysis of variance

    --->Anova helps us compare multiple groups and chek whether their average values are significantly different from each other

```python
from scipy.stats import f_oneway
method_a=[70,72,68,75,74]
method_b=[80,82,78,85,81]
method_c=[90,92,88,95,91]
#f_stat tells us whether the grps are significantly different from each other
f_stat,p_value=f_oneway(method_a,method_b,method_c)
print("f-statistic:",f_stat)
print("p-value:",p_value)
if p_value<0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
from scipy.stats import chisquare
observed = [20,10,5,5]
expected = [10,10,10,10]
chi_stat,p_value=chisquare(observed,expected)
print("chi-statistic:",chi_stat)
print("p-value:",p_value)
if p_value<0.05:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```
