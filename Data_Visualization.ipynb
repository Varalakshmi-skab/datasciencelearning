```python
import matplotlib.pyplot as plt
plt.plot([10,20,30,40,50,60])
plt.show()
x=[0,1,2,3,4,5]
y=[10,20,30,40,50,60]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7]
y=[5,20,60,50,50,50,100]
plt.plot(x,y)
plt.show()
```

```python
days=[1,2,3,4,5,]
sales=[100,200,5,500,700]
plt.figure(figsize=(5,5)) #(width,height) by using figure we can draw our graph in particular width and height
plt.title("Tirupati Sales Report")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.plot(days,sales)
plt.show()
```

```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5]
temperatures=[30,32,35,34,36]
plt.figure(figsize=(8,4))
plt.plot(days,temperatures)
plt.title("Temperature Analysis")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```

```python
days=[1,2,3,4,5]
productasales=[100,200,200,300,400]
productbsales=[1000,2000,3000,4000,5000]
productcsales=[10,20,30,40,50]
plt.plot(days,productasales,label="Product A")
plt.plot(days,productbsales,label="Product B")
plt.plot(days,productcsales,label="Product C")
plt.title("Sales Analysis")
plt.xlabel("Day")
plt.ylabel("Sales")
plt.legend()
plt.show()
```
