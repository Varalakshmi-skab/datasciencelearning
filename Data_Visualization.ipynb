#**BAR CHART**

*  A BAR CHART IS USED FOR COMPARING CATEGORIES
*   COMPARE SALES OF PRODUCTS
*   COMPARE STUDNET MARKS
*   COMPARE DEPARTMENT PLACEMENTS





```python
import pandas as pd
data = {
    "Product":["laptop","Mobile","Tablet","Monitor","Printer"],
    "Sales_2024":[250,300,350,400,500],
    "Sales_2025":[300,400,500,600,700]

}
df = pd.DataFrame(data)
print(df)
```

```python
plt.bar(
    df["Product"],
    df["Sales_2024"]
)
plt.show()
```

```python
plt.bar(
    df["Product"],
    df["Sales_2024"],
    color="green"
)
plt.show()
```

```python
plt.barh(
    df["Product"],
    df["Sales_2024"],
    color="green"
)
plt.show()
```

#why grouped bars?

*   we use grouped bars suppose if we want to Compare



```python
df.plot(
    x="Product",
    y = ["Sales_2024","Sales_2025"],
    kind = "bar"
)
plt.show()
```

#stacked bar chart put one bar on top of another

```python
 #stacked bar chart put one bar on top of another
plt.bar(
   df["Product"],
   df["Sales_2024"],
   color="green",
   label=2024
)
plt.bar(
    df["Product"],
    df["Sales_2025"],
    bottom= df["Sales_2024"],
    color="red",
    label=2024
)
plt.show()
```

#**Histogram**

  --->A histogram is used to show the distribution of numerical data
  
  --->it groups values into ranges (called bins) and counts how many values fall into each region

```python
# 50,60,60,50,48,49,50,51,52,52,71,67,68,90
# 40-50 => ?
# 50-60=>?
# 60-70=?
# 70-80=>?
marks = [45,50,46,47,40,49,50,55,65,70,71,72,81,60,80,91,92,93,81]
plt.hist(marks, bins=5)
plt.show()
```

#**seaborn**

   --->seaborn is a high level data visualization library build on top of matplotlib that helps  create attractive and informative statistical garphs with less code

   --->seaborn => plot creation

   --->matplot => title, labels, grid, figure

```python
import seaborn as sns
import matplotlib.pyplot as plt
flights = sns.load_dataset("flights")
flights
```

```python
sns.lineplot(
    data=flights,
    x ="year",
    y="passengers",
    errorbar= None
)
plt.show()
```

```python
sns.lineplot(
    data=flights,
    x ="year",
    y="passengers",

)
plt.title("data")
plt.show()
```

```python
from numpy import square
sns.lineplot(
    data=flights,
    x ="year",
    y="passengers",
    marker ="o"
   )
plt.title("data")
plt.show()
# o => circle
# s => square
# * => star
# d =>diamond
sns.lineplot(
    data=flights,
    x ="year",
    y="passengers",
   color="red")
plt.title("data")
plt.show()
# blue, red,orange, purple
sns.lineplot(
    data=flights,
    x ="year",
    y="passengers",
    linestyle="--"
 )
plt.title("data")
plt.show()
# -, --,:,
```

#**Heatmap**

   --->a heat map is a graphical representation of data where values are represented by colors

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
data = pd.DataFrame(
    [
        [95,90,80],
        [70,80,89],
        [10,20,30]
    ],
    columns = ["nath","scinece","physics"],
    index = ["A", "B","C"]

)
```

```python
sns.heatmap(data, cmap="Reds", )
plt.show()
```
