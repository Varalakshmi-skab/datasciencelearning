#**Groupby** : it enables catogary analysis instead of row by row analysis

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets for datascience/weather_by_cities.csv")
```

```python
df
```

```python
g=df.groupby("city")
```

```python
for city,city_df in g:
  print(city)
  print(city_df)
```

```python
g.get_group("mumbai")
```

```python
df.columns
```

```python
df.groupby("city")["temperature"].max()
```

```python
#find the avg wind speed per city
df.groupby("city")["windspeed"].mean()
```
