#**Encoding** : converting the text data into numerical data is called **Encoding**

--> Models can not understand our text data,it can only understand numerical data.

-->we must convert the text into numerical data,this process is called Encoding


-->Types of Encoding

* **Label Encoding** : it will assign one no. to each category

    -->This method assigns a unique integer to each category.

* **One-Hot Encoding** : it is a technique used to convert text data into numerical data by creating seperate columns for each category and filling them with 0s and 1s

    --> 1 : category present
    
    --> 0 : category not present


New library : ***scikit*** is an open sorce python library used for machine learning tasks such as data preprocessing, model building, model training, evaluation

##**Label Encoder**👇👇

```python
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
df=sns.load_dataset("tips")
```

```python
df
```

```python
#we use 1d data for Label encoder
sex_data=df["sex"]
```

```python
encoder=LabelEncoder()
encoded=encoder.fit_transform(sex_data)
#fit=seperate catogories
#transform=give numbers
```

```python
df["sex"]=encoded
```

##**One-Hot Encoding** 👇👇

```python
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
df=sns.load_dataset("tips")
df
```

```python
#we use 2d data for OneHotEncoder
x=df[["sex"]]
x
```

```python
encoder=OneHotEncoder(sparse_output= False)
encoded=encoder.fit_transform(x)
print(encoded)
```

```python
encoder.get_feature_names_out()
```

```python
encoded_df=pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out()
)
encoded_df
```

```python
d=df.drop("sex",axis=1)
d
```

```python
final_df = pd.concat([d,encoded_df],axis=1)
final_df
```

##**Task 9**

```python
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df=sns.load_dataset("diamonds")
df.head()
```

```python
print(df.describe())
print(df.head())
print(df.columns)
print(df.shape)
```

```python
cut_data = df[['cut']]
encoder = OneHotEncoder(sparse_output=False)
cut_encoded = encoder.fit_transform(cut_data)
cut_df = pd.DataFrame(cut_encoded, columns=encoder.get_feature_names_out(['cut']))
print(cut_df.head())
```

```python
color_data = df[['color']]
color_encoded = encoder.fit_transform(color_data)
color_df = pd.DataFrame(color_encoded, columns=encoder.get_feature_names_out(['color']))
print(color_df.head())
```

```python
clarity_data = df[['clarity']]
clarity_encoded = encoder.fit_transform(clarity_data)
clarity_df = pd.DataFrame(clarity_encoded, columns=encoder.get_feature_names_out(['clarity']))
print(clarity_df.head())
```

```python
df_final = df.drop(columns=['cut','color','clarity'])
final_df = pd.concat([df_final, cut_df, color_df, clarity_df], axis=1)
print(final_df.head())
print(final_df.shape)
print(final_df.columns.tolist())
```
