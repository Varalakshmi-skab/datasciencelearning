```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/Datasets for datascience/insurance_data.csv")
df.head()
```

```python
df.isnull().sum()
df.duplicated()
```

```python
df.corr()
```

```python
plt.scatter(df.age,df.bought_insurance)
plt.xlabel('Age')
plt.ylabel('Bought Insurance')
plt.title('Age vs Bought Insurance')
plt.show()
```

```python
X=df[["age"]]
Y=df["bought_insurance"]
print(X)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(
    X,
    Y,
    test_size=0.2,
    random_state=42
)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
print("Accuracy")
print(accuracy_score(Y_test,y_pred))
print("Confusion Matrix")
print(confusion_matrix(Y_test,y_pred))
print("Precision Score")
print(precision_score(Y_test,y_pred))
print("Recall Score")
print(recall_score(Y_test,y_pred))
print("F1 Score")
print(f1_score(Y_test,y_pred))
```

```python
import pickle
pickle.dump(model,open("model.pkl","wb"))
```

```python
import pickle
import gradio as gr
model=pickle.load(open("/content/drive/MyDrive/Datasets for datascience/model.pkl","rb"))
def predict_insurance(age):
  pred=model.predict([[age]])[0]
  if pred==1:
    return "They will buy insurance"
  else:
    return "They will not buy insurance"
app=gr.Interface(
    fn=predict_insurance,
    inputs=gr.Number(label="Age"),
    outputs=gr.Textbox(label="Prediction"),
    title="Insurance Prediction Website",
    description="Enter the age to predict whether they will buy insurance or not"
)
app.launch(share=True)
```
